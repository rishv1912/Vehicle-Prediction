### Converting data into True & False 

We there are string in the data, so we need to convert them into 1 & 0 for values.

- Fuel_Type
    - Petrol : 0  
    - Diesel : 1
- Seller_Type
    - Dealer : 0 
    - Individual : 1
- Transmission
    - Manual : 0
    - Automatic : 1


```python
# data["Fuel_type"]["Petrol"]=0
# data["Fuel_type"]["Diesel"]=1

# data["Seller_type"]["Dealer"]=0
# data["Seller_type"]["Invidual"]=1

# data["Transmission"]["Manual"]=0
# data["Transmission"]["Automatic"]=1

# for i in data["Fuel_Type"]:
#     if i == "Petrol":
#         data[i] = 0
#     if i == "Diesel":
#         data[i] 


data['Fuel_Type'] = data['Fuel_Type'].map({'Petrol': 0, 'Diesel': 1})
print("data['Fuel_type']",data["Fuel_Type"])

data['Seller_Type'] = data['Seller_Type'].map({'Dealer': 0, 'Individual': 1})
print("data['Seller_Type']",data["Seller_Type"])

data['Transmission'] = data['Fuel_Type'].map({'Manual': 0, 'Automatic': 1})
print("data['Transmission']",data["Transmission"])

```
